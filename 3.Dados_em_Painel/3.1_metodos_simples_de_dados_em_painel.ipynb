### Exemplo 13.1 - Fertilidade feminina ao longo do tempo

Questão de exemplo a ser repondida: após ter controlado a variável educação, o padrão de fertilidade entre mulheres com mias de 35 anos mudou entre 1972 e 1984? O seguinte exemplo ilustra como essa questão pode ser respondida de maneira mais simples, com o uso da análise de regressão múltipla com **variáveis dummy anuais**

```python
# Importando os dados
fertil1 = woo.dataWoo('fertil1')

# primeiras observações
fertil1.head()
```

```python
# Formato dos dados
#fertil1.shape
```

```python
# Dados faltantes por coluna
#fertil1.isna().sum()
```

Neste exemplo, o ano base é 1972. O coeficiente associado a y84 indica que, em comparação a 1972, no ano de 1984, as mulheres tiveram, em média, 0,54 filhos a menos. Isso significa que, mantendo todas as outras variáveis constantes, como educação, idade e outros fatores, um grupo de 100 mulheres em 1984 teria, em média, 54 filhos a menos do que um grupo semelhante de 100 mulheres em 1972.

```python
# Variável dependente
y = fertil1["kids"]

# Variáveis independentes
X = fertil1[[
    "educ", "age", "agesq", "black", "east", "northcen", "west",
    "farm", "othrural", "town", "smcity", "y74", "y76", "y78", "y80", "y82", "y84"
]]

# Adicionando uma constante
X = sm.add_constant(X)

# Ajusta o MQO
model = sm.OLS(y, X).fit()

# Exibe os resultados
print(model.summary())
```

### Exemplo 13.3 - Efeito da localização de um incinerador de Lixo sobre os preços de imóveis

Exemplo a seguir: Estudo de um evento mostra como dois conjuntos de dados de cortes transversais, coletados antes e depois da ocorrência de um evento, podem ser usados para determinar seu  efeito sobre resultados econômicos. Kiel e McClain (1995) estudaram o efeito que um novo incinerador de lixo teve sobre os valore dos imóveis em North Andover, Massachusetts.

```python
# Importando os dados
kielmc = woo.dataWoo('kielmc')

kielmc.head()
```

```python
# Verificando se há dados faltantes
#kielmc.isna().sum()
```

Começaremos verificando o efeito monetário sobre os preços dos imóveis. Uma análise ingênua usaria somente os dados de 1981 e estimaria um modelo muito simples

```python
kielmc81 = kielmc[kielmc['year'] == 1981]
```

```python
# Variável dependente
y = kielmc81["rprice"]

# Variáveis independentes
X = kielmc81['nearinc']

# Adicionando uma constante
X = sm.add_constant(X)

# Ajusta o MQO
model = sm.OLS(y, X).fit()

# Exibe os resultados
print(model.summary())
```

```python
# Exibe os coeficientes e erros padrão com formatação não científica
print("Coeficientes:")
for param, coef in zip(model.params.index, model.params.values):
    print(f"{param}: {coef:,.2f}")
```

A regressão acima não implica que a localização do incinerador esteja causando a queda nos valores dos imóveis. Na verdade, se rodarmos a mesma regressão para 1978 (antes mesmo de o incinerador ser cogitado), observamos que antes mesmo da instalação do incinerador no local definido os preços das casas já era menor nesta localidade.

```python
kielmc78 = kielmc[kielmc['year'] == 1978]
```

```python
# Variável dependente
y = kielmc78["rprice"]

# Variáveis independentes
X = kielmc78['nearinc']

# Adicionando uma constante
X = sm.add_constant(X)

# Ajusta o MQO
model = sm.OLS(y, X).fit()

# Exibe os resultados
print(model.summary())
```

```python
# Exibe os coeficientes e erros padrão com formatação não científica
print("Coeficientes:")
for param, coef in zip(model.params.index, model.params.values):
    print(f"{param}: {coef:,.2f}")
```

Como, então, podemos determinar se a construção de um novo incinerador reduz os valores dos imóveis? A chave é observar como o coeficiente de nearinc mudou entre 1978 e 1981. A diferença no valor médio das casas foi muito maior em 1981 do que em 1978 (US 30.688,27 contra US 18.824,37), mesmo em termos percentuais em relação ao valor médio das casas distantes do local do incinerador. A diferença entre os dois coeficientes de nearinc é.

$$ \hat\delta_1 = -30.668,27 - (-18.824,37) = - 11.863,9$$

Essa é a estimativa do efeito do incinerador sobre os valores das casas próximas ao local do incinerador. Na economia aplicada, ficou conhecido como o, $ \hat\delta_1$, estimador de diferença em diferenças (DD ou DID) porque pode ser expresso como:

$$ \hat{\delta}_{1}=(\overline{rprice}_{81,nr}-\overline{price}_{81,fr})-(\overline{rprice}_{78,nr}-\overline{rprice}_{78,fr}) $$

Para testar $ \hat\delta_1$ se é estatisticamente diferente de zero, precisamos encontrar seu erro padrão utilizando uma análise de regressão. Na verdade, $ \hat\delta_1$, pode ser obtido estimando

$$ rprice = \beta_0 + \delta_0 \gamma_{81} + \beta_1 nearinc + \delta_1 \gamma_{81} \cdot nearinc + u $$

Estimando a regressão acima, que tmabém é a coluna (1) da tabela 13.2

```python
# Variável dependente
y = kielmc["rprice"]

# Variáveis independentes
X = kielmc[['y81', 'nearinc', 'y81nrinc']]

# Adicionando uma constante
X = sm.add_constant(X)

# Ajusta o MQO
model = sm.OLS(y, X).fit()

# Exibe os resultados
print(model.summary())
```

```python
# Exibe os coeficientes e erros padrão com formatação não científica
print("Coeficientes:")
for param, coef in zip(model.params.index, model.params.values):
    print(f"{param}: {coef:,.2f}")
```

```python
# Variável dependente
y = kielmc["rprice"]

# Variáveis independentes
X = kielmc[['y81', 'nearinc', 'y81nrinc', 'age', 'agesq']]

# Adicionando uma constante
X = sm.add_constant(X)

# Ajusta o MQO
model = sm.OLS(y, X).fit()

# Exibe os resultados
print(model.summary())
```

Abaixo, coluna (3) da tabela 13.2

```python
# Variável dependente
y = kielmc["rprice"]

# Variáveis independentes
X = kielmc[['y81', 'nearinc', 'y81nrinc', 'age', 'agesq', 'intst', 'land', 'area', 'rooms', 'baths']]

# Adicionando uma constante
X = sm.add_constant(X)

# Ajusta o MQO
model = sm.OLS(y, X).fit()

# Exibe os resultados
print(model.summary())
```

```python
# Exibe os coeficientes e erros padrão com formatação não científica
print("Coeficientes:")
for param, coef in zip(model.params.index, model.params.values):
    print(f"{param}: {coef:,.2f}")
```

Agora, $ 100 . \delta_1 $ representa a redução percentual aproximada no valor das residências devido ao incinerador.

```python
# Variável dependente
y = kielmc["lprice"]

# Variáveis independentes
X = kielmc[['y81', 'nearinc', 'y81nrinc']]

# Adicionando uma constante
X = sm.add_constant(X)

# Ajusta o MQO
model = sm.OLS(y, X).fit()

# Exibe os resultados
print(model.summary())
```

O coeficiente do termo de interação implica que, devido ao novo incinerador, as casas próximas ao incinerador perderam cerca de 6,3% do seu valor. No entanto, essa estimativa não é estatisticamente diferente de zero. Mas quando usamos um conjunto completo de controles, como na coluna (3) da Tabela 13.2 (mas com intst, land e area aparecendo na forma logarítmica), o coeficiente de y81·nearinc se torna 2,132 com uma estatística t de cerca de 22,53. Novamente, controlar outros fatores se mostra importante. Usando a forma logarítmica, estimamos que as casas próximas ao incinerador foram desvalorizadas em cerca de 13,2%.

### Exemplo 13.9 - Taxas de criminalidade municipais na Carolina do Norte

```python
# Importando os dados
crime4 = woo.dataWoo("crime4")
crime4.head()
```

```python
# Verificando se há dados faltantes
#crime4.isna().sum()
```

```python
import numpy as np
import linearmodels as plm

crime4 = crime4.set_index(['county', 'year'], drop=False)

# estimate FD model:
reg = plm.FirstDifferenceOLS.from_formula(
    formula='np.log(crmrte) ~ year + d83 + d84 + d85 + d86 + d87 +'
            'lprbarr + lprbconv + lprbpris + lavgsen + lpolpc',
    data=crime4)
results = reg.fit()
print(f'results: \n{results}\n')
```
